2-Aplicar as regras de tratamento necessárias para preparar os dados de acordo com as especificações do banco de dados, use o Python para isso.

```python
dados = pd.DataFrame(var_dados)

dados = dados.astype({'Endereco':'string', 'CEP':'string','Cidade':'string','UF':'string','Nome':'string',})
#Converte dados para STRING

dados.info()
```

```python
dados['Nome'] = dados['Nome'].str.title()
dados
#Converte dados para .title()
```

```python
def valida_nome(nome):
    if any(char.isdigit() for char in str(nome)):
        return "Dados Inválidos"
    return nome

# Aplica a função na coluna "Nome"
dados['Nome'] = dados['Nome'].apply(valida_nome)
#Tira a opção de incluir numeros no campo Nome

dados
```

```python
#Valida entrada de dados númericos.
def valida_valor(valor):
    try:
        float(valor)
        return valor
    except ValueError:
        return "Dados Inválidos"

# Aplica a validação em cada coluna numérica
colunas_numericas = ['ID', 'Numero Paciente', 'RG', 'Numero do Prontuario']
for coluna in colunas_numericas:
    dados[coluna] = dados[coluna].apply(valida_valor)
```

```python
#Valida a entrada de dados da coluna Nome que precisam ser obrigatórios
dados['Nome'] = dados['Nome'].fillna("Informação obrigatória")
```

```python


# valida e formata a data para o formato DD/MM/AAAA
def formatar_data(data):
    try:
        # Coverte datas quando possível para o tipo datetime
        data_formatada = pd.to_datetime(data, errors='raise')
        # Retorna a data formatada no formato desejado DD/MM/AAAA
        return data_formatada.strftime('%d/%m/%Y')
    except Exception:
        # Quando não é possível, retorna a mensagem "Dados Inválidos"
        return 'Dados Inválidos'

# Aplicando a função na coluna 'Data de Nascimento'
dados['Data de Nascimento'] = dados['Data de Nascimento'].apply(formatar_data)

# Exibindo o DataFrame com as datas formatadas ou com a mensagem de erro
print(dados['Data de Nascimento'])
```

```python
#Valida formatação e validação do CEP
def valida_cep(cep):
    # Remove o traço ("-")
    cep_limpo = str(cep).replace('-', '')
    
    # Verifica se o valor contém apenas dígitos e tem exatamente 8 caracteres
    if cep_limpo.isdigit() and len(cep_limpo) == 8:
        return cep_limpo
    else:
        return "Dados Inválidos"

# Aplica a função de validação na coluna 'CEP'
dados['CEP'] = dados['CEP'].apply(valida_cep)

print(dados['CEP'])

```

```python
#Substitui todos os (NaN) para informação vazia.
dados = dados.fillna('')
```

```python
#valida e formata regras de Data de Nascimento.

# Converte as colunas para o formato datetime.
dados['Data de Nascimento'] = pd.to_datetime(dados['Data de Nascimento'], format='%d/%m/%Y', errors='coerce')
dados['Data de Nascimento Responsavel'] = pd.to_datetime(dados['Data de Nascimento Responsavel'], format='%d/%m/%Y', errors='coerce')

# Cria uma máscara para identificar as linhas com inconsistência temporal.
mask_invalida = dados['Data de Nascimento Responsavel'] > dados['Data de Nascimento']

# Converte as datas para string no formato desejado.
dados['Data de Nascimento'] = dados['Data de Nascimento'].dt.strftime('%d/%m/%Y').fillna('Dados Inválidos')
dados['Data de Nascimento Responsavel'] = dados['Data de Nascimento Responsavel'].dt.strftime('%d/%m/%Y').fillna('Dados Inválidos')

# Nas linhas inválidas, substitui os valores das colunas pela mensagem "Dados Inválidos".
dados.loc[mask_invalida, 'Data de Nascimento'] = 'Dados Inválidos'
dados.loc[mask_invalida, 'Data de Nascimento Responsavel'] = 'Dados Inválidos'

# Exibe apenas as colunas de interesse para conferência.
print(dados[['Data de Nascimento', 'Data de Nascimento Responsavel']])


```

```python
#Validação e aplicação de regas para Telefone e Celular


def validar_formatar_numero(num):
    # Se o campo estiver vazio, retorna vazio
    if pd.isnull(num) or str(num).strip() == "":
        return ""
    
    # Remove todos os caracteres não numéricos
    num_digits = re.sub(r'\D', '', str(num))
    
    # Se após remover os não dígitos o campo ficar vazio, retorna inválido
    if num_digits == "":
        return "Dados Inválido"
    
    # Verifica se o campo contém somente números
    if not num_digits.isdigit():
        return "Dados Inválido"
    
    # Valida se a quantidade de dígitos está correta (13 dígitos, incluindo o código do país)
    if len(num_digits) != 13:
        return "Dados Inválido"
    
    # Remove os dois primeiros dígitos (código do país)
    numero_sem_cod = num_digits[2:]
    
    # Após remover, o número deve ter 11 dígitos
    if len(numero_sem_cod) != 11:
        return "Dados Inválido"
    
    # O DDD é composto pelos dois primeiros dígitos; verificação extra: 
    # ele deve ser composto de dois dígitos e não pode ser "00"
    ddd = numero_sem_cod[:2]
    if not ddd or len(ddd) != 2 or ddd == "00":
        return "Dados Inválido"
    
    # Aplica a formatação (XX)XXXXX-XXXX
    parte1 = numero_sem_cod[2:7]  # Próximos 5 dígitos
    parte2 = numero_sem_cod[7:]   # Últimos 4 dígitos
    return f"({ddd}){parte1}-{parte2}"


# Aplica a função nas colunas 'Celular' e 'Telefone'
for coluna in ['Celular', 'Telefone']:
    dados[coluna] = dados[coluna].apply(validar_formatar_numero)

```

```python
# Exibe apenas as colunas 'Celular' e 'Telefone' para verificação
print(dados[['Celular', 'Telefone']])
```

```python
#Validar formatação CPF 

def validate_cpf(cpf: str) -> bool:
    # Remove qualquer pontuação e mantém apenas os números
    numbers = [int(digit) for digit in cpf if digit.isdigit()]

    # Verifica se o CPF possui 11 números ou se todos são iguais:
    if len(numbers) != 11 or len(set(numbers)) == 1:
        return False

    # Validação do primeiro dígito verificador:
    sum_of_products = sum(a * b for a, b in zip(numbers[0:9], range(10, 1, -1)))
    expected_digit = (sum_of_products * 10 % 11) % 10
    if numbers[9] != expected_digit:
        return False

    # Validação do segundo dígito verificador:
    sum_of_products = sum(a * b for a, b in zip(numbers[0:10], range(11, 1, -1)))
    expected_digit = (sum_of_products * 10 % 11) % 10
    if numbers[10] != expected_digit:
        return False

    return True

```

```python

# Aplica a validação do CPF
dados['CPF'] = dados['CPF'].apply(lambda cpf: cpf if validate_cpf(cpf) else 'Dados Inválidos')

# Cria uma máscara para os CPFs que não são 'Dados Inválidos'
mask_valid = dados['CPF'] != 'Dados Inválidos'

# Cria uma máscara para linhas onde o CPF aparece mais de uma vez
mask_duplicated = dados.duplicated(subset=['CPF'], keep=False)

# Remove as linhas onde o CPF é válido e está duplicado
dados = dados[~(mask_valid & mask_duplicated)]

```

```python
#Validar UF 

# Função para validar o UF: deve conter exatamente duas letras (maiúsculas ou minúsculas)
def validate_uf(uf):
    # Se não for string, retorna False
    if not isinstance(uf, str):
        return False
    # Verifica se o UF contém exatamente duas letras e nada mais
    return bool(re.fullmatch(r'[A-Za-z]{2}', uf))

# Aplica a validação na coluna 'UF':
# Se o valor for válido, mantém; caso contrário, substitui por "Dados Inválidos"
dados['UF'] = dados['UF'].apply(lambda uf: uf if validate_uf(uf) else 'Dados Inválidos')

# Exibe as primeiras linhas para conferência
print(dados[['UF']])

```

```python
#Validação e-mail 


def validate_email(email):
    # Verifica se o e-mail é do tipo string
    if not isinstance(email, str):
        return False
    # aplica o padrão esperado 
    padrao = r'^[\w\.-]+@servico\.com(\.br)?$'
    return bool(re.fullmatch(padrao, email))


# Se o e-mail for válido, mantém o valor original; caso contrário, substitui por "Dados Inválidos".
dados['E-mail'] = dados['E-mail'].apply(lambda email: email if validate_email(email) else 'Dados Inválidos')


print(dados[['E-mail']])
```
